# Question 1

Occupations<br>
Step 1. Import the necessary libraries<br>
Step 2. Import the dataset from the provided github address<br>
Step 3. Assign it to a variable called users<br>
Step 4. Discover what is the mean age per occupation<br>
Step 5. Discover the Male ratio per occupation and sort it from the most to the least<br>
Step 6. For each occupation, calculate the minimum and maximum ages<br>
Step 7. For each combination of occupation and sex, calculate the mean age<br>
Step 8. For each occupation present the percentage of women and men<br>

### Steps 1, 2, and 3

```python
import pandas as pd

users = pd.read_csv(r"https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep="|")
users.head()
```

### Step 4: Discover what is the mean age per occupation

```python
meanGB = pd.Series(users.groupby(['occupation'])['age'].mean())
meanAge = pd.DataFrame({'Mean Age':meanGB})
meanAge
```

### Step 5: Discover the male ratio per occupation and sort it from most to least

```python
occgen = users.groupby(['occupation', 'gender']).count()
piv = pd.pivot_table(occgen,'user_id','occupation','gender')
piv['F'] = piv['F'].fillna(0)
piv['% Ratio M'] = ((100/(piv['F'] + piv['M']))*piv['M'])
df2 = piv.sort_values('% Ratio M',ascending=False).loc[:]['% Ratio M']
display(pd.DataFrame(df2))
```

### Step 6: For each occupation, calculate the minimum and maximum ages

```python
minAge = pd.Series(users.groupby(['occupation'])['age'].min())
maxAge = pd.Series(users.groupby(['occupation'])['age'].max())
minMax = pd.DataFrame({'Min Age':minAge, 'Max Age':maxAge})
minMax
```

### Step 7: For each combination of occupation and sex, calculate the mean age

```python
meanGB = pd.Series(users.groupby(['occupation','gender'])['age'].mean())
meanAge = pd.DataFrame({'Mean Age':meanGB})
piv = pd.pivot_table(meanAge, index='occupation', columns='gender')
display(pd.DataFrame(piv))
```

### Step 8: For each occupation present the percentage of women and men

```python
occgen = users.groupby(['occupation', 'gender']).count()
piv = pd.pivot_table(occgen,'user_id','occupation','gender')
piv['F'] = piv['F'].fillna(0)
piv['% Ratio F'] = ((100/(piv['F'] + piv['M']))*piv['F'])
piv['% Ratio M'] = ((100/(piv['F'] + piv['M']))*piv['M'])
#df2 = piv.sort_values('% Ratio M',ascending=False).loc[:]['% Ratio M','% Ratio F']
display(pd.DataFrame(piv))
```
